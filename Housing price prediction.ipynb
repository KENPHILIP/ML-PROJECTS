```python
# supervised ml uses label data and outcomes and has a feaures and atarget
# unsupervised ml uses non labelled data
# our problem requires supervised ml
```

```python
# Algorithms for supervised ml
#     Linear Regression,
#     Logistic Regression,
#     Support Vector Machine, 
#     Multi-class Classification, 
#     Decision tree,
#     Bayesian Logic
# Algorithms for unsupervised ml
    # Clustering, 
    # KNN
    # Apriori algorithm.
```

```python
housingdata.columns
```

```python
y=housingdata['Price']
```

```python
x=housingdata.drop(['Price','Address'], axis =1)
```

```python
x
```

```python
x.head(10)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
our_model=LinearRegression()
our_model.fit(x_train,y_train)
print(our_model)
```

```python
import pickle
```

```python
saved_model= pickle.dumps(our_model)
```

```python
pickle.dump(our_model, open("this_model.pkl","wb"))
```

```python
# Evaluate model
```

```python
preds=our_model.predict(x_test)
print(preds)
```

```python
plt.scatter(y_test,preds)
```

```python
from sklearn.metrics import r2_score
r2_score(preds,y_test)
```
